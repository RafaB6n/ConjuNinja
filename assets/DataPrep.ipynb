```python
# Import the pandas library  
import pandas as pd  
  
# Define the path to your CSV file  
csv_file_path = 'jehle_verb_database.csv'  # Please replace with your actual file path  
  
# Use the read_csv function from pandas to read the CSV file  
df = pd.read_csv(csv_file_path)
```

```python
df.head(5)
```

```python
# Create an empty dictionary to store the final result  
verbs_dict = {}
cur_id = 0

# Iterate over each row in the dataframe  
for _, row in df.iterrows():  
    # Concatenate 'tense' and 'mood' to form the inner dictionary's key  
    inner_key = f'{row["tense"]} {row["mood"]}'
    
    # Extract the verb forms as list  
    forms = [row[col] if row[col] else '' for col in df.columns if col.startswith('form_')]  
    forms = [f if type(f) is str else '' for f in forms]
    # Get the infinitive for the current row  
    infinitive = row['infinitive']
  
    # Add this data to the verbs dictionary  
    if infinitive not in verbs_dict:
        verbs_dict[infinitive] = {
            inner_key: forms,
            '_id': cur_id,
            'infinitive_english': row['infinitive_english'],
            'pastparticiple_english': row['pastparticiple_english']
        }
        cur_id += 1
        
    else:  
        verbs_dict[infinitive][inner_key] = forms
```

```python
from collections import defaultdict

count = defaultdict(int)
for verb, conj in verbs_dict.items():
    count[len(conj)] += 1
print(count)
```

# Id to Conjugation Map

```python
# Create an empty dictionary to store the final result  
id_dict = {}
cur_id = -1
prev_infinitivo = None

# Iterate over each row in the dataframe  
for _, row in df.iterrows():  
    # Concatenate 'tense' and 'mood' to form the inner dictionary's key  
    inner_key = f'{row["tense"]} {row["mood"]}'
    
    # Extract the verb forms as list  
    forms = [row[col] if row[col] else '' for col in df.columns if col.startswith('form_')]  
    forms = [f if type(f) is str else '' for f in forms]
    # Get the infinitive for the current row  
    infinitivo = row['infinitive']
  
    # Add this data to the verbs dictionary  
    if infinitivo != prev_infinitivo:
        cur_id += 1
        id_dict[cur_id] = {
            inner_key: forms,
            'Inifitivo': infinitivo,
            'infinitive_english': row['infinitive_english'],
            'pastparticiple_english': row['pastparticiple_english']
        }
    else:
        id_dict[cur_id][inner_key] = forms
    
    prev_infinitivo = infinitivo
```

```python
import json  
  
# specify the file name  
json_file = 'id_dict.json'  
  
# use the 'dump' method from the 'json' module to write the dictionary to a file  
with open(json_file, 'w') as file:  
    json.dump(id_dict, file)
```
